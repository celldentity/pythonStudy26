```python
# .readlines() s가 하나 더 붙는데 메모장 파일을 리스트로 한 번에 불러온다.
# 파일의 모든 줄을 읽어 리스트 형태로 반환
# 반환값은 문자열 리스트 ['첫번째줄\n', '두번째줄\n',,,,,,]
# 파일 내용을 모두 메모리에 로드하기 때문에 파일크기가 클경우 메모리 부족!
# 장점 : 파일의 모든 줄을 한번에 가져와 처리해야 할 때 편리
```

```python
# 메모장에 엔터효과인 \n 을 지우고 싶으면 strip을 쓰면 된다.
# strip은 문자열내에서 원하는 문자열 또는 공백을 제거함
# lstrip() 원하는 문자열 또는 공백을 왼쪽에서 지우는 거
# rstrip() 원하는 문자열 또는 공백을 오른쪽에서 지우는 거
# ^strip() 원하는 문자열 또는 공백을 왼쪽&오른쪽 모두에서 지우는 거
f = open("c:/mbc/새파일.txt", "r")
lines = f.readlines()
for line in lines:
    line = line.strip() # 줄끔에 줄 바꿈 문자를 제거할 수 있다. 엔터도 아랫줄의 공백 하나와 같다.
    print(line)
f.close()
```

```python
# 파일에 새로운 내용을 추가 append -> a
# w로 파일을 열면 이미 존재하는 파일이면 그 내용이 사라지고 덮어쓰기
# a모드는 원래 있던 값을 유지하면서 새로운 값을 맨 아래에 추가!!
f = open("c:/mbc/새파일.txt", "a")  # append (추가)
f. write("--------------------------")
for i in range(11, 20):
    data = "%d번째 줄입니다. \n" % i
    f.write(data)
f.close()
# 회원관리를 파일처리 해보자
```
