```python
# 숫자형(numberType) : 숫자로 이루어진 자료형으로
# 123과 같은 정수, 12.34와 같은 실수, 8진법, 16진법 등이다.

# 정수형(integer) : 양의정수, 음의정수, 0
a = 123
print(a)
```

```python
b = -123
print(b) # 음의 정수
```

```python
# 실수형(floating-point) : 소수점이 포함된 숫자
c = 1.2
print(c) # 양의 실수
```

```python
d = -3.45
print(d) # 음의 실수
```

```python
# 지수형식 (과학기술용) 엄청큰 실수 들~!
e = 4.24e10 # 42.400.000.000
print(e)
```

```python
f = 4.24e-10 # 0.00000000424
print(f)
```

```python
# 8진수(octal) 입력 기법
# 숫자0과 영문자o를 쓰고 입력한다.
g = 0o177 # 8진법 177
print(g)
```

```python
# 16진법(hexadecimal) 0x로 입력
h = 0x8ff  # 16진법 8ff
print(h)
```

```python
# 사칙연산
# a+b 더하기
# a-b 빼기
# a*b 곱하기
# a**b  제곱근 a의 b승
4**4
```

```python
3#a/b 나누기
5/2
```

```python
#a//b 나머지를 제외한 몫 구하기
5//2
```

```python
5%2 #나머지
```

```python
# 복합 연산자
# 산술연산자와 대입연산자를 합쳐 놓은 것
i = 0
i = i + 1 # i변수에 1을 더하고 i에 넣어라 '나를 계속 변화시켜라'
print(i)
```

```python
i += 1   # i = i + 1 을 줄인형태
print(i)
```

```python
i -= 1  # i = i - 1 을 줄인형태
print(i)
```

```python
i *= 5  # i = i * 5
print(i)
```
