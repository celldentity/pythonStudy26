```python
# 주민번호를 입력 받아 생년월일 남녀 구분을 하는 코드
# input() 함수를 사용하면 콘솔로 데이터를 넣을 수 있다.

print("주민번호를 입력하세요!!(-포함 14글자)")
ssn = input(">>>")
print("입력된주민번호 : "+ssn)
year = ssn[0:2]
month = ssn[2:4]
day = ssn[4:6]
sex = int(ssn[7])

if sex % 2 == 0:
    gender = "여자"
else:
    gender = "남자"

if sex == 1:
    yearif = "19"+year
if sex == 2:
    yearif = "19"+year
if sex == 5:
    yearif = "19"+year
if sex == 6:
    yearif = "19"+year
if sex == 9:
    print("당신은 외계인입니다. 돌아가세요!!")
else:
    yearif = "20"+year

print("당신의 생년은 " + year + "년")
print("당신의 생월은 " + month + "월")
print("당신의 생일은 " + day + "일")
print("당신은 " + yearif + "대생 입니다.")
```
