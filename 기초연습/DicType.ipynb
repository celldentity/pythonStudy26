```python
# 딕셔너리 타입은 key : value로 이루어진다.
# {key1 : value1, key2 : value2 ....}
# 대부분 웹 프로그램 언어는 json 이라는 값으로 처리 하는데
# 유사한 형태이다.

dic = {"name" : "kkw", "age" : 25}
#        k        v      k      v
type(dic)
print(dic["name"])
```

```python
a = {1 : 'hi'}
print(a[1])
#  배열번호는 key 번호다
```

```python
b = {"a" : [1, 2, 3]}
#키를 만들고 대응하는 리스트를 밸류로 넣을 수도 있다.
print(b["a"])
```

```python
#만약 b에서 2라는 걸 출력하고 싶다면
print(b["a"][1]) # 이렇게 원하는 배열을 찾을 때는 딕셔너리롤 활용할 수 있다.
```

```python
# get을 이용해 키를 넣으면 벨류가 나오도록 하는 법
dic.get("name")
```

```python
# key에 해당하는 값을 가져올 때 없는 키를 넣으면 에러발생 []방식으로 작동하는 것보다 예외처리를 해서 계속해서 런 상태를만들어 놓는 게 중요하다. 프로그램은 안꺼지고 오류메세지를 띄우는거다.
# 예외처리는 다음에 배우게 될 거다.
dic["age"] # 없는 키를 가져오려고 하면 오류가 발생한다.
# 프로그램 실행중 치명적인 오류로 강제 종료 된다. 강제 종료되지 않도록 하는 게 중요하다.
#(인서트 키가 눌려진 상황에서는 엔터가 입력되지 않는다. 수정용이다.)
#---------------------------------------------------------------------------
#KeyError                                  Traceback (most recent call last)
#Cell In[49], line 2
#      1 # key에 해당하는 값을 가져올 때 없는 키를 넣으면 에러발생
#----> 2 dic["age"]
#
#KeyError: 'age'
```

```python
dic.get("age", "값 없음")
# 키에 해당하는 벨류가 없고, 키도 없는데도 불구하고 없으면 그냥 아무런 답을 안띄우고 넘어간다. 치명적인 오류를 띄우질 않는다.
# 일반적으로get을 쓰면 모르는 값이 들어가도 오류가 나질 않는다. (그냥 none 이라고 프롬프트 상에는 나온다.)
# get은 첫번쨰가 Key이고 key가 없을 떄의 출력문으로 응용해서 사용할 수 있다. 딕셔너리는 겟이랑 궁합이 좋다.
```

```python
print(dic)
```

```python
"name" in dic #in 명령어는 리스트나, 튜플, 딕셔너리 안에 값이 있는지를 묻는 것 트루 반환한다.
```

```python
"김기원" in dic  #in 함수로는 딕셔너리 안에서 key 밖에 찾아내질 못한다.
```

```python
# key를 이용해서 value - pop 방법이 또 있다.
print(dic)
```

```python
email = dic.pop("email") # 값을 꺼내와 변수에 넣고 삭제한다.
print(email)
```
