```python
# 튜플(Tuple)은 리스트와 유사하지만 수정과 삭제가 불가능한 배열
# 리스트는 []로 작성하고 튜플은 ()로 작성한다.
# 1개의 값만 튜플에 넣을 때는 ㅁ = (1,)마지막에 쉼표가 필수다.
# b = 1, 2, 3, 4 처럼 입력해도 튜플 처리 된다.
```

```python
# 튜플 생성은 변수에 넣은 값을 활용 하는 용도
t1 = (1, 2, 'a', 'b')
print(t1)
```

```python
type(t1)
```

```python
# 값을 수정하면 오류가 발생한다.
t1[3] = 8 #튜플을 생성할때만 무조건 소괄호로 작성하고 인덱싱 번호는 항상 대괄호다.
```

```python
# 값을 삭제해도 오류가 발생한다
del t1[2]
```
