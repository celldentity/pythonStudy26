```python
### 문자열 포매팅
print("1. I eat %d apples." % 3)       # % 숫자 -> 숫자 바로 대입 '문자열 포맷 코드' 일종
print("2. I eat %s apples." % "five")  # % 문자열 바로 대입

number = 3
print("3. I eat %d apples." % number)

number = number + 7
day = "three"
print("4. I ate %d apples. so I was sick for %s days" % (number, day)) #2개 이상의 포매팅을 할 땐 %뒤에 괄호로 묶어줘야 한다.

# %s	문자열(String)
# %c	문자 1개(character)
# %d	정수(Integer)
# %f	부동소수(floating-point)
# %o	8진수
# %x	16진수
# %%	Literal % (문자 % 자체)

print("5. rate is %s" % 3.234)  # 문자열로 입력할 경우 정수나 소수와 상관없이 입력해 넣을 수 있다.
print("6. Error is %d%%." % 98) # %를 나타내려면 반드시 포매팅 연산자 뒤에 %%을 두개 입력해야 한다. 포매팅 연산자가 없으면 뒤에 % 혼자 쓰여도 괜찮다.
print("7. %10s" % "hi") # 전체 길이가 10개인 문자열 공간에서 대입되는 값을 오른쪽으로 정렬하고 나머지는 공백으로 남겨두라는 의미다.
print("8. %-10sjane." % "hi") # 반대의 의미

print("9. %0.4f" % 3.42134234) # 0은 전체 문자열의 길이, 전부 가져온다는 뜻으로 생략하기도 한다. 소수점 뒤 4는 소수점 아래로 몇 자리까지 가져올지를 표시한다.
print("10. %10.4f" % 3.42134234) # 소수점 네 번째 자리까지만 표시하고 전체 길이가 10개인 문자열 공간에서 오른쪽으로 정렬하는 예를 보여줌

#format 함수를 사용한 포매팅
print("11. I eat {0} apples.".format(3))  # 숫자 바로 대입하기
print("12. I eat {0} apples.".format("five")) # 문자열 바로 대입하기

number1 = 4936
print("13. I eat {0} apples.".format(number1)) #  문자열의 항목을 사전에 설정한 변수값으로 대입할 수 있다.

number2 = 29
day1 = "four"
print("14. I eat {0} apples. so I was sick for {1} days".format(number2, day1)) #2개 이상의 값을 넣을 경우, 문자열의 {0}, {1}과 같은 인덱스 항목이 format 함수의 입력값으로 순서에 맞게 바뀐다. 위 예에서 {0}은 format 함수의 첫 번째 입력값인 number2, {1}은 format 함수의 두 번째 입력값인 day1으로 바뀌었다.

print("15. I eat {number3} apples. so I was sick for {day2} days".format(number3=20, day2=5)) # 인덱스 항목 대신에 더 편리한 {name}형태를 사용하는 방법도 있다. 이 형태를 사용하게 될 경우, format 함수에는 반드시 name=value와 같은 형태의 입력값이 있어야 한다.

print("16. I ate {0} apples. so I was sick for {day3} days.".format(15, day3=2)) #인덱스 항목과 네임벨류를 함께 사용하는 것도 가능.
print("17. {0:<10}".format("hi")) #치환되는 문자열을 왼쪽으로 정렬하고 문자열의 총 자릿수를 10으로 맞출 수 있다.
print("18. {0:>10}".format("dude")) #반대방향이다. :<, :> 둘 중에 하나를 사용해서 정렬방향을 결정할 수 있다.
print("19. {0:^10}".format("maduro")) #가운데 정렬도 가능하다.
print("20. {0:=^10}".format("Trump"))
print("21. You are {0:!<10}!!!".format("fired")) #정렬할 때 공백 문자 대신 지정한 문자 값으로 채워 넣을 수도 있다. 채워 넣을 문자 값은 정렬 문자 <, >, ^ 바로 앞에 넣어야 한다. 위 예에서 첫 번째 예제는 가운데(^)로 정렬하고 빈 공간을 =로 채웠고, 두 번째 예제는 왼쪽(<)으로 정렬하고 빈 공간을 !로 채웠다.

y = 3.42134234
print("22. {0:0.4f}".format(y)) # 숫자를 가져올건데, 포매이션이 y로 소수를 가져왔다.
print("23. {0:10.4f}".format(y)) # 자릿수를 10개로 맞추고 소수점 자리 4개까지

print("24. {{ and }}".format()) # {}와 같은 중괄호 문자를 문자 그대로 사용하고 싶은 경우 2개를 연속해서 사용하면 된다.

name = '김정은'
age = 42
print("25." + f'나의 이름은 {name}입니다. 나이는 {age}입니다.')
print("26." + f'정은이는 내년이면 {age+1}살이 된다.')
#format 이라는 단어로 적기가 너무 귀찮아서 그냥 f로 시작하고 따옴표가 있으면 포매팅으로 인식하도록 만들어졌다.

#딕셔너리는 f 문자열 포매팅에서 다음과 같이 사용할 수 있다.
F = {'name':'트럼프', 'age':81}
print("27." + f'나의 이름은 {F["name"]}입니다. 나이는 {F["age"]}입니다.')

print("28." + f'{"what is this":<20}')  # 왼쪽 정렬)
print("29." + f'{"what is this":>20}')  # 오른쪽 정렬)
print("30." + f'{"what is this":^20}')  # 가운데 정렬)

print("31." + f'{"hello":=^20}') # 가운데 정렬하고 '=' 문자로 공백 채우기
print("32." + f'{"wtf":!<20}') # 왼쪽 정렬하고 '!' 문자로 공백 채우기

y = 3.421342341564
print("33." + f'{y:0.4f}') # 소수점 4자리까지만 표현
print("34." + f'{y:8.4f}') # 소수점 4자리까지 표현하고 총 자리수를 8로 맞춤
print("35." + f'{{where are we}}') # 중괄호를 그대로 표시하려면 2개를 동시에 사용해야 한다.

print("36." + f"난 {280000000:,}원이 필요해") # 콤마가 없이 적혀있는 부분은 f 문자열 콤마를 활용하여 금액에 표시할 수 있다.
```

```python
### 문자열 관련 함수들

## 문자 개수 세기 = count
h = "do it your self"
hc = h.count('t')
print("1. " + str(hc)) # count 함수를 통해 문자열 중 문자 t의 개수를 반환했다.

## 포함된 문자 위치 알려주기 1 - find
i = "Python is the best choice"
iif = i.find('b')
iiff = i.find('k')
print("2. " + str(iif)) # find 함수로 문자열 중 문자 b가 처음 나온 위치를 반환했다. 만약 찾는 문자나 문자열이 존재하지 않는다면 -1을 반환한다.
print("3. " + str(iiff))

## 포함된 문자 위치 알려주기 2 - index
j = "Life is too short"
jc = j.index('t')
jcc = j.count('k')
print("4. " + str(jc)) #index 함수로 문자열 중 문자 t가 맨 처음으로 나온 위치를 반환했다.
print("5. " + str(jcc)) #index 함수는 만약 찾는 문자나 문자여링 존재하지 않는다면 오류가 발생한다.

## 문자열 삽입 = join
print("6. " + ",".join('abcd')) #join 함수로 abcd 문자열의 각각의 문자 사이에 '.'를 삽입했다.
# join 함수는 문자열뿐만 아니라 앞으로 배울 리스트나 튜플도 입력으로 사용할 수 있다.
print("7. " + ",".join(['a', 'b', 'c', 'd']))

## 소문자를 대문자로 바꾸기 - upper
a = "hi"
print("8. " + a.upper())

## 대문자를 소문자로 바꾸기 - lower
a = "HI"
print("9. " + a.lower())

## 왼쪽 공백 지우기 - rstrip
a = "  hi  "
print("10. " + a.lstrip())

## 오른쪽 공백 지우기
print("11. " + a.rstrip())

## 양쪽 공백 지우기
print("12. " + a.strip())

## 문자열 바꾸기 - replace
a = "Life is too short"
print("13. " + a.replace("Life", "Your leg"))

## 문자열 나누기 = split
a = "Life is too short"
print("14. " + str(a.split()))

## 문자열이 알파벳으로만 구성되어 있는지 확인하기 - isalpja
s = "destination"
print("15. " + str(s.isalpha()))
s1 = "crypto3"
print("16. " + str(s1.isalpha()))
s2 = "Hello World Best Driver"
print("17. " + str(s2.isalpha()))
# 이즈알파 함수는 문자열이 알파벳 문자로만 이루어져 있는지 검사한다. 영문 알파벳만 포함되어 있는 경우에만 True를 반환하며, 공백이나 숫자, 특수 문자가 포함되면 False를 반환한다.

## 문자열이 숫자로만 구성되어 있는지 확인하기 - isdigit
l = "12345"
print("18. " + str(l.isdigit()))
l1 = "12345a"
print("19. " + str(l1.isdigit()))
l2 = "12  34"
print("20. " + str(l2.isdigit()))
# isdigit 함수는 문자열이 숫자(0~9)로만 이루어져 있는지를 검사한다. 문자열의 모든 문자가 숫자일 경우 True를 반환하고, 하나라도 숫자가 아닌 문자가 포함되어 있으면 False를 반환한다.

## 문자열이 특정 문자 (열)로 시작하는지 확인하기 - startswith
k = "Life is too short"
print("21. " + str(k.startswith("Life")))
print("22. " + str(k.startswith("short")))

## 문자여링 특정 문자(열)로 끝나는지 확인하기 - endswith
print("23. " + str(k.endswith("short")))
print("24. " + str(k.endswith("Life")))

## 착각하기 쉬운 문자열 함수
# upper 뿐만 아니라 lower, join, lstrip, rstrip, strip, replace, split 함수는 모두 이와 같은 규칙이 적용되어 문자열 자체의 값이 변경되는 것이 아니라 변경된 값을 반환한다.
# 항상 변경하려는 변수를 앞에 두고 . upper, lower, join, lstrip, rstrip, strip, replace, split 가 입력되어야 한다.
```
