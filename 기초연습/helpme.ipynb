```python
run = True
menu = """
===========================
MBC 아카데미 회원 관리 프로그램
===========================
1. 회원가입
2. 로그인
3. 회원보기
4. 내정보수정
5. 프로그램종료
"""

# 데이터 저장 리스트
sns = [1, 2]
ids = ["kkw", "lhj"]
passwords = ["1234", "4321"]
names = ["관리자", "임효정"]
emails = ["admin@mbc.com", "lhj@mbc.com"]
admins = [True, False]

# 로그인 세션 변수 (None이면 로그아웃 상태, 숫자면 해당 유저의 인덱스)
idx = None

while run:
    print(menu)
    # 로그인 상태 표시
    if login_idx is not None:
        print(f" 현재 접속 중: {names[idx]}님")

    select = input("1~5 숫자를 입력하여 메뉴를 선택하세요 : ")

    if select == "1":
        print("\n[회원가입 메뉴]")
        # ... (기존 회원가입 로직과 동일)
        sn = input("사번을 입력하세요 : ")
        id = input("아이디를 입력하세요 : ")
        pw = input("암호를 입력하세요 : ")
        name = input("이름을 입력하세요 : ")
        email = input("이메일주소를 입력하세요 : ")

        print(f"입력 확인 - 이름: {name}, ID: {id}")
        if input("가입하시겠습니까? (y/n) : ").lower() == "y":
            sns.append(sn); ids.append(id); passwords.append(pw)
            names.append(name); emails.append(email); admins.append(False)
            print("회원가입이 완료되었습니다.")

    elif select == "2":
        print("\n[로그인 메뉴]")

        # 로그인 성공할 때까지 반복하는 루프
        while True:
            input_id = input("아이디 (취소: 'q'): ")
            if input_id.lower() == 'q':
                break # 로그인 취소 후 메인 메뉴로 이동

            input_pw = input("비밀번호 : ")

            if input_id in ids:
                idx = ids.index(input_id)
                if passwords[idx] == input_pw:
                    login_idx = idx # 로그인 성공 index 저장
                    print(f"\n✅ {names[idx]}님 로그인 성공!")
                    if admins[idx]:
                        print("▶ 관리자 권한으로 접속되었습니다.")
                    break # 성공했으므로 반복문 탈출
                else:
                    print("❌ 비밀번호가 틀렸습니다. 다시 시도하세요.")
            else:
                print("❌ 존재하지 않는 ID 입니다. 다시 입력하세요.")

    elif select == "3":
        if login_idx is None:
            print("로그인이 필요한 서비스입니다.")
            continue

        print("\n[회원보기]")
        # 관리자일 경우 전체 리스트, 일반 유저일 경우 본인 정보만
        if admins[login_idx]:
            print("--- 전체 회원 리스트 (관리자 전용) ---")
            for i in range(len(ids)):
                print(f"이름: {names[i]} | ID: {ids[i]} | 관리자여부: {admins[i]}")
        else:
            print(f"내 정보 -> 이름: {names[login_idx]}, 이메일: {emails[login_idx]}")

    elif select == "5":
        print("프로그램을 종료합니다.")
        run = False
```
