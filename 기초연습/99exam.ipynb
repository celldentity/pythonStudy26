```python
# 구구단을 만들어보자

# 구구단은 앞단과 뒷단을 곱해서 출력하면됨!
# i      j
# I = 1 2 3 4 5 6 7 8 9     -반복적으로 실행 for문
#      J = 1 2 3 4 5 6 7 8 9     -반복적으로 실행 for문
# I * J = Print

# 1~9까지 반복코드를 만드는 방법은?
# i = i + 1     ->9 이상이면 종료! 이건 while문 쓸 때는 이렇게
# range()    내장된 함수를 사용하게 되면 for문을 쓴다.
# range(9)
for i in range(9):
    print(i, end=" ")  # for문 반복마다 출력한다. (end는 출력 뒤에 붙이는 코드)
```

```python
for i in range(1, 10):   #1에서 10이되기 전까지 반복한다. 이게 뒷단이다.
    print(i, end=" ")
```

```python
# for 를 이용한 range활용 구구단

print("=====구구단 출력 용 코드=====")
for i in range (2,10) : #2부터 9까지 반복
    print(f"-------{i}단---------")
    for j in range(1,10) : #뒷단 1부터 9까지 반복
        print(f"{i} x {j} = {i * j}")
```

```python
# 만약 키보드로 앞단을 입력 받으면 뒷단을 1~9까지 자동 계산한다.

def backDan(i):
    print(f"========={i}단을 출력합니다.==========")
    for j in range(1, 10): # 1부터 9까지 반복용
        print(f"{i} x {j} = {i * j}")
```

```python
i = int(input("확인하려는 구구단 앞단을 입력하세요 :"))
backDan(i)
```
