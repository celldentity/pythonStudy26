```python
# 함수를 만든다. (함수정의)
def add(a, b): #add 라는 함수를 만들고 인구는 a와 b를 받는다.
    return a + b #받은 a와 b를 더하고 돌려준다.
```

```python
# 위에서 만든 활용해보자
c = add(5, 10) # 위에서 만든 add 함수에 5와 10을 넣고 결과를 c에 넣는다.
print(c)
# 초보자들의 실수는 매개변수에 값을 넣지 않는 거다. 디파인에서 정의를 했더라도 실제로 사용할 때 받을 변수가 무조건 있어야 한다.
#
```

```python
# 만약 함수에 입력값이 몇 개가 될지 모를때는?
# args는 인수라는 뜻으로 다른 글자를 사용해도 됨!
# 그러나 프로그래머들의 관례로 대부분 args로 사용함
def add_many(*args): # *을 붙이면 몇 개의 값이 들어오든~ 이라는 뜻이다.
    result = 0  #리절트에 0인데 숫자 취급을 받았지만 뒤에 텍스트가 들어와서 문제가 된다.
    for i in args: #튜플로 인수가 들어와서
        result = result + i #튜플로 되어 있는 값이 번갈아 더함
    return result
```

```python
# add_many 테스트
print(add_many(1, 2, 3))
```

```python
resultText = add_many("안녕하세요", "반갑습니다", "파이썬 지옥입니다.")
print(resultText)
```

```python
#리턴 없이 글로벌 변수를 사용하고 싶은 경우 이걸 통해서 함수를 사용해서 계속 만든다.
session = "로그인안함"    #쌍따옴표는 문자열처리, None 숫자열 처리
def login(id):
    """
    키보드로 입력받은 문자열을
    리스트에서 찾아 중복 검사를 진행하고
    없으면 회원가입을 할 수 있도록 처리

    작성자 : 김수빈
    :param id: 문자열
    :return: 로그인 상태를 세션에 저장한다.(글로벌)
    """
    global session   # 글로벌 명령어는 전역변수를 가져와!!!!
    session = id     # 글로벌 명령어로 같은 변수명은 사용 금지

login(input("아이디를 입력하세요 : "))
print(login.__doc__)
```
