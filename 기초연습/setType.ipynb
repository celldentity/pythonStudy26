```python
# 초기 데이터 예시
users = ["User1", "User2", "User3"]  # 일반 사용자 리스트 (가입 순서대로라고 가정)
admins = []  # 현재 관리자 리스트 (비어있음)

def check_and_promote_admin(user_list, admin_list):
    # 1. 관리자 리스트가 비어있는지 확인
    if not admin_list:
        print("경고: 관리자가 없습니다. 시스템이 자동으로 권한을 부여합니다.")

        # 2. 일반 사용자 리스트에 사람이 있는지 확인
        if user_list:
            # 3. 인덱스상 가장 빠른 번호(0번)를 추출하여 관리자로 승격
            new_admin = user_list.pop(0)
            admin_list.append(new_admin)
            print(f"결과: '{new_admin}'님이 새로운 관리자가 되었습니다.")
        else:
            print("오류: 권한을 부여할 사용자가 시스템에 존재하지 않습니다.")
    else:
        print(f"현재 관리자 목록: {admin_list}")

# 실행
check_and_promote_admin(users, admins)

print(f"업데이트된 관리자: {admins}")
print(f"남은 사용자: {users}")
```

```python
scores = []
print("10개의 성적을 입력받습니다 (0~100점).")

while len(scores) < 10:
    val = input(f"{len(scores) + 1}번째 성적 입력: ")

    try:
        # 1. 숫자로 변환 시도 (문자 입력 시 여기서 에러 발생)
        num = int(val)

        # 2. 범위 체크 (0~100점)
        if 0 <= num <= 100:
            scores.append(num)
        else:
            print("경고: 0에서 100 사이의 점수만 입력 가능합니다.")

    except ValueError:
        # 숫자가 아닌 문자가 들어왔을 때 실행되는 구간
        print("에러: 문자는 입력할 수 없습니다. 숫자만 입력해주세요.")

print("입력 완료된 성적:", scores)
```

```python
import time
import sys

def typewriter(text, speed=0.05):
    for char in text:
        sys.stdout.write(char) # 한 글자씩 출력
        sys.stdout.flush()     # 버퍼를 비워 즉시 화면에 표시
        time.sleep(speed)      # 다음 글자까지 대기 시간 (초 단위)
    print() # 마지막에 줄바꿈

# 사용 예시
typewriter("만나서 반갑습니다 저는 김수빈이라고 합니다 이런 타입 라이터를 사용하면 훨씬 아날로그합니다")
```
